```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly

import h5py # to read in the hdf5 files

from scipy.fftpack import fft, ifft # for FFT analysis
```

```python
DATA_PATH = '../../../data/zenodo'
```

```python
sub_dirs = ['01-Start 3V', '02-Change 5V', '03-Re_start 3V', '04-Change 5V']
```

```python
os.listdir('../../../data/zenodo')
```

## 01-Start 3V

```python
fnames = os.listdir(os.path.join(DATA_PATH, sub_dirs[0]))
```

```python
print(f'Number of files: {len(fnames)}')
```

```python
d0 = h5py.File(os.path.join(DATA_PATH, sub_dirs[0], 'MOTOR-DC_2020_12_02_17_31_00_Analogico.hdf5'), 'r')
data0 = d0['open_DC_motor'][:]
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
plt.plot(data0[:, 0], '-', linewidth=2, label='Vibration')
plt.plot(data0[:, 1], '-', linewidth=2, label='Current')
plt.plot(data0[:, 2], '-', linewidth=2, label='Voltage')
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
plt.plot(data0[1:500, 0], '-', linewidth=2, label='Vibration')
plt.plot(data0[1:500, 1], '-', linewidth=2, label='Current')
plt.plot(data0[1:500, 2], '-', linewidth=2, label='Voltage')
plt.legend()
```

### Vibration Spectrum

```python
x = data0[:, 0]

if(len(x)%2 !=0):
    x = x[:-1]

N = len(x)

# Sample spacing
T = 1.0 / 51200
time_points = np.linspace(0.0, N*T, N)

xf = np.linspace(0.0, 1.0/(2.0*T), int(N/2))

yf = fft(x)


fig, ax = plt.subplots(figsize=(12, 6))
#plt.plot(xf, 2.0/N * np.abs(yf[0:int(N/2)]))
plt.plot(xf[:2000], 2.0/N * np.abs(yf[0:int(N/2)][:2000]))
plt.xlabel('Frequency (Hz)', fontsize=14)
plt.ylabel('Amplitude', fontsize=14)
```

### Current Spectrum

```python
x = data0[:, 1]

if(len(x)%2 !=0):
    x = x[:-1]

N = len(x)

# Sample spacing
T = 1.0 / 51200
time_points = np.linspace(0.0, N*T, N)

xf = np.linspace(0.0, 1.0/(2.0*T), int(N/2))

yf = fft(x)


fig, ax = plt.subplots(figsize=(12, 6))
#plt.plot(xf, 2.0/N * np.abs(yf[0:int(N/2)]), 'orange')
plt.plot(xf[:2000], 2.0/N * np.abs(yf[0:int(N/2)][:2000]), 'orange')
plt.xlabel('Frequency (Hz)', fontsize=14)
plt.ylabel('Amplitude', fontsize=14)
```

### Voltage Spectrum

```python
x = data0[:, 2]

if(len(x)%2 !=0):
    x = x[:-1]

N = len(x)

# Sample spacing
T = 1.0 / 51200
time_points = np.linspace(0.0, N*T, N)

xf = np.linspace(0.0, 1.0/(2.0*T), int(N/2))

yf = fft(x)

fig, ax = plt.subplots(figsize=(12, 6))
#plt.plot(xf, 2.0/N * np.abs(yf[0:int(N/2)]), 'green')
plt.plot(xf[:2000], 2.0/N * np.abs(yf[0:int(N/2)][:2000]), 'green')
plt.xlabel('Frequency (Hz)', fontsize=14)
plt.ylabel('Amplitude', fontsize=14)
```
